```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the data source directory
source_data_dir = '../Source data/Extended Data Figure 6'
npz_path = os.path.join(source_data_dir, "stats_plot_data.npz")

# Load the NPZ file
data = np.load(npz_path, allow_pickle=True)["plot_data"].item()

# Define colors for each region
region_colors = {"region1": "blue", "region2": "red"}

# Iterate through each plot name and generate the corresponding figure
for plot_name, plot_info in data.items():
    lims = plot_info["lims"]  # Set y-axis limits

    # Create a two-row subplot: CA1 (region1) on top, M1 (region2) at the bottom
    fig, (ax1, ax2) = plt.subplots(
        2, 1, figsize=(6, 2.3),
        sharex=True,  # Share x-axis
        gridspec_kw={'height_ratios': [1, 1], 'hspace': 0}  # Adjust spacing between subplots
    )

    # ======== Plot CA1 (region1) Data ========
    for subject, subject_data in plot_info["region1"].items():
        ax1.scatter(subject_data["lapses"], subject_data["means"], color=region_colors["region1"], alpha=0.3)
        ax1.plot(subject_data["lapses"], subject_data["means"], color=region_colors["region1"], alpha=0.3)

    # Overlay the mean trend using seaborn's pointplot
    sns.pointplot(
        x=[x for subj in plot_info["region1"].values() for x in subj["lapses"]],
        y=[y for subj in plot_info["region1"].values() for y in subj["means"]],
        capsize=0.2, color=region_colors["region1"], 
        scale=0.5, ax=ax1  # Fixed the error
    )

    # Customize y-axis for CA1
    ax1.set_ylabel(f'{plot_name} (CA1)', color=region_colors["region1"])
    ax1.tick_params(axis='y', labelcolor=region_colors["region1"], direction='in')
    ax1.tick_params(axis='x', which='both', bottom=False, top=False)  # Hide x-axis labels
    ax1.spines['bottom'].set_visible(False)  # Remove bottom border line
    ax1.set_ylim(lims[0], lims[1])  # Set y-axis limits

    # ======== Plot M1 (region2) Data ========
    for subject, subject_data in plot_info["region2"].items():
        ax2.scatter(subject_data["lapses"], subject_data["means"], color=region_colors["region2"], alpha=0.3)
        ax2.plot(subject_data["lapses"], subject_data["means"], color=region_colors["region2"], alpha=0.3)

    # Overlay the mean trend using seaborn's pointplot
    sns.pointplot(
        x=[x for subj in plot_info["region2"].values() for x in subj["lapses"]],
        y=[y for subj in plot_info["region2"].values() for y in subj["means"]],
        capsize=0.2, color=region_colors["region2"], 
        scale=0.5, ax=ax2  # Fixed the error
    )

    # Customize y-axis for M1
    ax2.set_ylabel(f'{plot_name} (M1)', color=region_colors["region2"], labelpad=10)
    ax2.yaxis.set_label_position('right')  # Move y-axis label to the right
    ax2.yaxis.tick_right()  # Move ticks to the right
    ax2.tick_params(axis='y', labelcolor=region_colors["region2"], direction='in')
    ax2.tick_params(axis='x', direction='in')  # Ensure x-axis ticks point inward
    ax2.spines['top'].set_visible(False)  # Remove top border line
    ax2.set_ylim(lims[0], lims[1])  # Set y-axis limits

    # Set shared x-axis label
    ax2.set_xlabel('Lapses')

    # Adjust layout to avoid overlapping elements
    fig.tight_layout()

    # Save the figure in the source data directory
    
    # plt.savefig(f'{source_data_dir}/{plot_name}_merged.pdf', transparent=True, bbox_inches='tight')
    
    # Display the plot in Jupyter Notebook
    plt.show()
    
    # Close the figure to free memory
    plt.close()

print(" Successfully redrawn all plots from NPZ data.")
```
